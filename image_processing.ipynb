Resizing of image using pillow

```python
from PIL import Image
```

```python
img=Image.open('/content/dog.jpg')
img_resize=img.resize((200,200))
```

```python
img_resize.save('dog_image_resized.jpg')
```

```python
#displaying an image using matplotlib image module
img_res=mping.imread('/content/dog_image_resized.jpg')
img_res_plot=plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

converting RGB IMGS TO GRAYSCALE IMGS USIJG OpenCV

```python
#importing OpenCV library
import cv2
```

```python
img=cv2.imread('/content/dog.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_img)
```

```python
grayscale_img.shape
```

cv2.imshow() will display image but will not work in google colab

```python
from google.colab.patches import cv2_imshow
```

```python
#display the converted image
cv2_imshow(grayscale_img)
```

```python
#saving grayscale img
cv2.imwrite('dog_grayscale_img.jpg',grayscale_img)
```
